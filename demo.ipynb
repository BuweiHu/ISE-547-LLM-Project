## Job Match Data

```python
df2 = pd.read_csv("raw_datasets/resume_data.csv")
df2.head()
```

```python
print(df2.columns.tolist())
```

```python
def safe_text(x):
    if pd.isna(x):
        return ""
    return str(x)

def parse_list(x):
    try:
        if pd.isna(x):
            return ""
        if isinstance(x, str) and x.startswith("["):
            return ", ".join(ast.literal_eval(x))
        return str(x)
    except:
        return str(x)

def build_resume_text(row):
    parts = [
        f"Career Objective: {safe_text(row['career_objective'])}",
        f"Skills: {parse_list(row['skills'])}",
        f"Education: {parse_list(row['degree_names'])} at {parse_list(row['educational_institution_name'])}",
        f"Major: {parse_list(row['major_field_of_studies'])}",
        f"Experience: {parse_list(row['professional_company_names'])}",
    ]
    return "\n".join([p for p in parts if p.strip() != ""])

def get_col(row, candidates):
    for c in candidates:
        if c in row.index:
            return row[c]
    return ""

def build_jd_text(row):
    job_title = get_col(row, ["job_position_name", "\ufeffjob_position_name", "job_position_name "])
    edu_req = get_col(row, ["educationaL_requirements", "educational_requirements"])
    exp_req = get_col(row, ["experiencere_requirement", "experience_requirement"])
    resp = get_col(row, ["responsibilities.1", "responsibilities"])
    skills_req = get_col(row, ["skills_required", "skills required"])

    parts = [
        f"Job Title: {safe_text(job_title)}",
        f"Education Requirement: {safe_text(edu_req)}",
        f"Experience Requirement: {safe_text(exp_req)}",
        f"Responsibilities: {safe_text(resp)}",
        f"Skills Required: {safe_text(skills_req)}",
    ]
    return "\n".join([p for p in parts if p.strip() != ""])


df2["resume_text"] = df2.apply(build_resume_text, axis=1)
df2["jd_text"] = df2.apply(build_jd_text, axis=1)

df_clean = df2[["resume_text", "jd_text", "matched_score"]]

df_clean = df_clean[
    (df_clean["resume_text"].str.len() > 50) &
    (df_clean["jd_text"].str.len() > 50)
]

processed_df2 = df_clean.sample(100, random_state=42)

processed_df2.to_csv("processed_dataset/processed_dataset2.csv", index=False)

print("Done! Saved to processed_dataset/processed_dataset2.csv")
```

### Metrics

```python
def get_clean_metrics_df(df):
    clean_df = df.dropna(subset=['actual_score', 'predicted_score']).copy()
    
    if clean_df['actual_score'].max() <= 1.0:
        clean_df['actual_score'] = clean_df['actual_score'] * 5.0

    def calculate_stats(group):
        y_true = group['actual_score']
        y_pred = group['predicted_score']
        
        mae = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        
        if len(y_true) > 1 and y_true.std() != 0 and y_pred.std() != 0:
            corr, _ = pearsonr(y_true, y_pred)
        else:
            corr = 0
            
        match_rate = (np.abs(y_true - y_pred) <= 0.5).mean() * 100
        
        return {
            'MAE': round(mae, 3),
            'RMSE': round(rmse, 3),
            'Correlation': round(corr, 3),
            'Match_Rate_Pct': round(match_rate, 2)
        }
    stats_list = []
    for (model, prompt), group in clean_df.groupby(['model', 'prompt_version']):
        metrics = calculate_stats(group)
        metrics['model'] = model
        metrics['prompt_version'] = prompt
        stats_list.append(metrics)
    
    metrics_df = pd.DataFrame(stats_list)
    
    cols = ['model', 'prompt_version', 'MAE', 'RMSE', 'Correlation', 'Match_Rate_Pct']
    return metrics_df[cols]

df = pd.read_csv("results/matching_results_final.csv")
df = df[df['model'] != "thm/glm-4-9b-chat:free"]

final_metrics_df = get_clean_metrics_df(df)

print("--- Final Metrics Table (Flat DataFrame) ---")
final_metrics_df

# final_metrics_df.to_csv("results/model_performance_metrics_flat.csv", index=False)
```

```python
plot_df = final_metrics_df.copy()

model_map = {
    "arcee-ai/trinity-large-preview:free": "Arcee Trinity (400B)",
    "openai/gpt-oss-120b:free": "GPT-OSS (120B)"
}
prompt_map = {
    "v1_baseline": "Baseline",
    "v2_expert": "Expert",
    "v3_cot": "CoT"
}
plot_df['model'] = plot_df['model'].replace(model_map)
plot_df['prompt_version'] = plot_df['prompt_version'].replace(prompt_map)

sns.set_theme(style="whitegrid", font_scale=1.2)

plt.figure(figsize=(10, 6))
ax1 = sns.barplot(
    data=plot_df, 
    x='prompt_version', 
    y='MAE', 
    hue='model', 
    order=['Baseline', 'Expert', 'CoT'],
    palette="RdYlGn_r"
)

plt.title('Prediction Error Analysis (MAE)', fontsize=18, pad=20)
plt.ylabel('Mean Absolute Error', fontsize=14)
plt.xlabel('Prompt Strategy Stage', fontsize=14)
plt.ylim(0, 3.0)

for p in ax1.patches:
    if p.get_height() > 0:
        ax1.annotate(format(p.get_height(), '.2f'), 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 9), textcoords='offset points', weight='bold')

plt.tight_layout()
plt.savefig('MAE_Final_Plot.png', dpi=300)
plt.show()


plt.figure(figsize=(10, 6))
ax2 = sns.barplot(
    data=plot_df, 
    x='prompt_version', 
    y='Match_Rate_Pct', 
    hue='model', 
    order=['Baseline', 'Expert', 'CoT'],
    palette="viridis"
)

plt.title('Candidate Match Reliability (±0.5 Margin)', fontsize=18, pad=20)
plt.ylabel('Match Rate (%)', fontsize=14)
plt.xlabel('Prompt Strategy Stage', fontsize=14)

for p in ax2.patches:
    if p.get_height() > 0:
        ax2.annotate(format(p.get_height(), '.1f') + '%', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 9), textcoords='offset points', weight='bold')

plt.tight_layout()
plt.savefig('Match_Rate_Final_Plot.png', dpi=300)
plt.show()
```
